```python
from pyzipcode import ZipCodeDatabase
zcbd = ZipCodeDatabase()
df = pd.read_csv('/content/kc_house_data.csv')

df
```

```python
df['date'] = pd.to_datetime(df['date'], format = '%Y%m%dT%H%M%S') #vamos ajustar a data que está em formato incorreto

df
```

```python
df['bathrooms'] = df['bathrooms'].astype(int) #vamos ajustar os valores dos banheiros 

df
```

```python
cities = []  #Vamos alinhar as cidades 
for zipcode in df['zipcode']:
  try:
    zc = zcbd[zipcode]
    cities.append(zc.city)
  except KeyError:
    cities.append(float('nan'))

df['city'] = cities
df
  
```

```python
df['id'] = df['date'].dt.month 
grouped = df.groupby("sqft_lot")["price"].mean()
median = df.groupby("sqft_lot")["price"].median()
std = df.groupby("sqft_lot")["price"].std()

df

```
