```python
# Adicionando o arquivo necessário 

from pyzipcode import ZipCodeDatabase
zcbd = ZipCodeDatabase()
df = pd.read_csv('/content/drive/MyDrive/kc_house_data.csv')

df
```

```python
# Ajuste de data que está em formato incorreto

df['date'] = pd.to_datetime(df['date'], format = '%Y%m%dT%H%M%S') 

df
```

```python
# Ajuste dos valores de banheiros 

df['bathrooms'] = df['bathrooms'].astype(int) 

df
```

```python
# Alinhamento das cidades 

cities = []  
for zipcode in df['zipcode']:
  try:
    zc = zcbd[zipcode]
    cities.append(zc.city)
  except KeyError:
    cities.append(float('nan'))

df['city'] = cities
df
  
```

```python
# Alinhamento dos ids, preço e sqft_lot, agrupando e realizando a mediana dos valores 

df['id'] = df['date'].dt.month 
grouped = df.groupby("sqft_lot")["price"].mean()
median = df.groupby("sqft_lot")["price"].median()
std = df.groupby("sqft_lot")["price"].std()

df

```

```python
 # Calculando o looping dentro do dataframe usando o index 1. Nos próximos códigos, é possível verificar os lucros e prejuízos por imóvel de forma mais sucinta 

for index in range (1,int(df.shape[0])):
  if df.iloc[index]['id'] == df.iloc[index-1]['id']:
     df.at[index,'valorization'] = df.iloc[index]['price'] -  df.iloc[index-1]['price']
     df.at[index,'hold_time'] = (df.iloc[index]['date']) - df.iloc[index-1]['date']
df

```

```python
valorization_areas = df.groupby('city')['valorization']\
                          .mean().reset_index()\
                          .sort_values('valorization', inplace=False)\
                          .reset_index(drop=True)
                        
```

```python
immobile_count = df.groupby('city')['id'].nunique().to_frame('immobile_count')
resale_info = pd.merge(valorization_areas, valorization_areas, on= 'city', how= 'inner')
resale_info = pd.merge(resale_info, immobile_count, on='city', how= 'inner')
```

```python
# Iniciando a importação de gráficos para distribuir os valores de preços, regiões onde se localizam as casa e a valorização por preço médio por cidade

import plotly.express as px
import plotly.graph_objs as go
resale_info_fig = df.plot.bar
```

```python
print(resale_info.columns)
```

```python
resale_info_fig2 = px.bar(resale_info, x='city', y='valorization_y', color='immobile_count', 
                         title= 'Preço médio de casa por cidade')

resale_info_fig2.add_trace(go.Scatter(x=resale_info['city'],
                                      y=resale_info['valorization_y'],
                                      name='ROI/Year',
                                      mode='lines'))


resale_info_fig2.update_layout(title_x=0.5,
                               template='simple_white',
                               yaxis_title='',
                               xaxis_title='',
                               hovermode="x unified")
resale_info_fig2.show()
df.plot
```
