```python
# Converter Date para datetime

df['Date'] = pd.to_datetime(df['Date'])

# Criando coluna para semana do ano

df['Week'] = df['Date'].dt.week
df
```

```python
# Buscando os tipos de feriados e suas datas, calculamos a mediana das vendas de cada loja antes de criamos uma coluna apenas com os eventos e quando ocorrem 

holidays = {'Super Bowl': ['2010-02-12', '2011-02-11', '2012-02-10'],
            'Labour Day': ['2010-09-10', '2011-09-09', '2012-09-07'],
            'Thanksgiving': ['2010-11-26', '2011-11-25', '2012-11-23'],
            'Christmas': ['2010-12-31', '2011-12-30', '2012-12-28']}


holiday_dates = []
for holiday in holidays.values():
    holiday_dates += holiday
holiday_dates = pd.to_datetime(holiday_dates)

df
            
```

```python
df['During_Holiday'] = df['Date'].isin(holiday_dates)

outlier_holidays = ['Super Bowl', 'Thanksgiving', 'Christmas']

df
```

```python
med_sales = df.groupby('Store')['Weekly_Sales'].median().reset_index()
med_sales.columns = ['Store', 'Median_Sales']

df = pd.merge(df, med_sales, on='Store', how='left')
```

```python
# Criando coluna com o nome dos respectivos feriados 

def Feriados(Data):
    if Data == "2010-02-12":
        return "Super Bowl"
    elif Data == "2011-02-11":
        return "Super Bowl"
    elif Data == "2012-02-10":
        return "Super Bowl"
    elif Data == "2010-07-10":
      return "Labour Day"
    elif Data == "2011-07-09":
      return "Labour Day"
    elif Data == "2012-07-07":
      return "Labour Day"
    elif Data == "2010-11-26":
        return "Thanksgiving"
    elif Data == "2011-11-25":
        return "Thanksgiving"
    elif Data == "2012-11-23":
      return "Thanksgiving"
    elif Data == 2010-12-31:
      return "Christmas"
    elif Data == 2011-12-31:
      return "Christmas"
    elif Data == "2010-5-3":
      return "Christmas"
    else:
        return 0



df["Nome_Feriado"] = df["Date"].apply(Feriados)

df
```

```python
# Criando uma função que recebe uma linha do dataframe e retorna o tipo de evento correspondente à data

def get_event_type(row):
    if row["Holiday_Flag"] == 1:
        if row["Date"].month == 2 and row["Date"].day >= 1 and row["Date"].day <= 7:
            return "Super Bowl"
        elif row["Date"].month == 9 and row["Date"].day >= 1 and row["Date"].day <= 7:
            return "Labour Day"
        elif row["Date"].month == 11 and row["Date"].day >= 22 and row["Date"].day <= 28:
            return "Thanksgiving"
        elif row["Date"].month == 12 and row["Date"].day >= 22 and row["Date"].day <= 28:
            return "Christmas"
    return "None"

df
```

```python
df["Event_Type"] = df.apply(get_event_type, axis=1)

df
```

```python
# Agrupando  por loja e semana e calcular a média das vendas semanais

weekly_sales_by_store = df.groupby(['Store', 'Week'])['Weekly_Sales'].mean().reset_index()

# Agrupando por loja e semana e calcular a soma das vendas semanais em feriados

holiday_sales_by_store = df[df['Holiday_Flag'] == 1].groupby(['Store', 'Week'])['Weekly_Sales'].sum().reset_index()
df
```

```python
# Mesclando as informações de vendas semanais e vendas em feriados
sales_by_store = pd.merge(weekly_sales_by_store, holiday_sales_by_store, on=['Store', 'Week'], how='left')
sales_by_store.rename(columns={'Weekly_Sales_x': 'Avg_Weekly_Sales', 'Weekly_Sales_y': 'Holiday_Sales'}, inplace=True)
sales_by_store.fillna(0, inplace=True)
sales_by_store['Total_Sales'] = sales_by_store['Avg_Weekly_Sales'] + sales_by_store['Holiday_Sales']

print(sales_by_store.head())
```

```python
# Criando  coluna para classificar os feriados

df.loc[(df['Date'].dt.month == 2) & (df['Date'].dt.day >= 1) & (df['Date'].dt.day <= 8), 'Holiday_Type'] = 'Super Bowl'
df.loc[(df['Date'].dt.month == 9) & (df['Date'].dt.weekday == 0) & (df['Date'].dt.day >= 1) & (df['Date'].dt.day <= 7), 'Holiday_Type'] = 'Labour Day'
df.loc[(df['Date'].dt.month == 11) & (df['Date'].dt.weekday == 4) & (df['Date'].dt.day >= 22) & (df['Date'].dt.day <= 28), 'Holiday_Type'] = 'Thanksgiving'
df.loc[(df['Date'].dt.month == 12) & (df['Date'].dt.day >= 1) & (df['Date'].dt.day <= 25), 'Holiday_Type'] = 'Christmas'

df
```

```python
# Substituindo valores de Holiday_Flag pelos tipos de feriado correspondentes

df.loc[(df['Holiday_Flag'] == 1) & (df['Holiday_Type'].isna()), 'Holiday_Type'] = 'Other'
df['Holiday_Flag'] = df['Holiday_Type']

# Agrupando por semana e calcular soma das vendas semanais, contagem de feriados e média de vendas em feriados

sales_by_week = df.groupby('Week').agg({'Weekly_Sales': 'sum', 'Holiday_Flag': 'count', 'Weekly_Sales': 'mean'}).reset_index()

df = df.dropna()

df
```

```python
# Calculando a soma de vendas por loja
df_soma_vendas = df.groupby('Store')['Weekly_Sales'].sum()

# Identificando as lojas com maior venda acumulada
lojas_maior_venda = df_soma_vendas.nlargest(5).index.tolist()

df
```

```python
# Calculando a média de vendas por semana para todo o período
media_vendas_semana = df['Weekly_Sales'].mean()

# Identificando as semanas em que cada loja ultrapassou a média de vendas por semana
df['Semana'] = pd.to_datetime(df['Date']).dt.week
df_media_vendas_semana = df.groupby(['Store', 'Semana'])['Weekly_Sales'].mean()
lojas_media_venda_acima = df_media_vendas_semana.reset_index().groupby('Store').apply(lambda x: (x['Weekly_Sales'] > media_vendas_semana).sum()).loc[lojas_maior_venda]

# Calculando o número de semanas que cada loja ultrapassou a média de vendas por semana
num_semanas_acima_media = lojas_media_venda_acima.tolist()

df
```

```python
# Imprimindo o resultado

for i in range(len(lojas_maior_venda)):
    print(f"A loja {lojas_maior_venda[i]} teve uma venda acumulada de {df_soma_vendas[lojas_maior_venda[i]]:.2f} e ultrapassou a média de vendas por semana em {num_semanas_acima_media[i]} semanas.")

df
```

```python
# Criando uma nova coluna 'Holiday' que indica se a data é um feriado ou não, com base na coluna 'Holiday_Flag'

df['Holiday'] = df['Holiday_Flag'].apply(lambda x: 'Holiday' if x == 'Yes' else 'Non-Holiday')

df
```

```python
# Criando uma nova coluna 'Holiday_Type' que indica qual é o tipo de feriado, com base na coluna 'Holiday_Flag'

df['Holiday_Type'] = df['Holiday_Flag'].apply(lambda x: 'Super Bowl' if x == 'Super Bowl' 
                                                            else ('Labour Day' if x == 'Labour Day' 
                                                                  else ('Thanksgiving' if x == 'Thanksgiving' 
                                                                        else ('Christmas' if x == 'Christmas' 
                                                                              else 'Non-Holiday'))))
df
```

```python
# Agrupando os dados por data e tipo de feriado, e calculando a venda máxima, mínima e média para cada grupo

grouped = df.groupby(['Date', 'Holiday_Type'])['Weekly_Sales'].agg(['max', 'min', 'mean']).reset_index()

df
```

```python
# Montando um gráfico de linha para visualizar as vendas máximas, mínimas e médias ao longo do tempo, com linhas separadas para cada tipo de feriado

fig, ax = plt.subplots(figsize=(10, 6))

for holiday in ['Super Bowl', 'Labour Day', 'Thanksgiving', 'Christmas']:
    data = grouped[grouped['Holiday_Type'] == holiday]
    ax.plot(data['Date'], data['max'], label=f'{holiday} Max')
    ax.plot(data['Date'], data['min'], label=f'{holiday} Min')
    ax.plot(data['Date'], data['mean'], label=f'{holiday} Mean')

ax.set_xlabel('Date')
ax.set_ylabel('Weekly Sales')
ax.set_title('Weekly Sales by Holiday')
ax.legend()

plt.show()
```

```python
# Datas das vendas que ocorreram dentro dos feriados

sales_df = df[["Store", "Date", "Holiday_Flag", "Event_Type"]]

def get_top_10_sales(event_type):
    event_sales = sales_df[sales_df["Event_Type"] == event_type]
    return event_sales.nlargest(10, "Sales")

sales_df = sales_df.groupby("Event_Type").apply(lambda x:(x.name))

df
```

```python
# Top 10 das lojas com maior percentual de crescimento e plot do gráfico

df['Taxa de crescimento'] = df.groupby('Store')['Weekly_Sales'].pct_change()

taxa_crescimento_media = (1 + df.groupby('Store')['Taxa de crescimento'].agg(np.mean))**(1/3) - 1
top_10_lojas = taxa_crescimento_media.nlargest(10).index.unique()

top_10_df = df[df['Store'].isin(top_10_lojas)]

taxa_crescimento_medio_por_loja = top_10_df.groupby('Store')['Taxa de crescimento'].mean()

x_labels = [str(loja) for loja in top_10_lojas]

x_pos = np.arange(len(top_10_lojas))
y_values = taxa_crescimento_media.loc[top_10_lojas]
plt.bar(x_pos, y_values)
plt.xticks(x_pos, x_labels)

plt.xlabel('ID da Loja')
plt.ylabel('Taxa de Crescimento Anual Média')

plt.show()
```
